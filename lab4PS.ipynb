2.2 2. Import necessary modules:

```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import load_iris
```

3 Step 2: Load and Explore the Dataset

3.1 1.Load the Iris dataset:

```python
iris = load_iris(as_frame=True  )
df = iris['frame']
df.head()
```

```python
df = df[['sepal length (cm)', 'petal length (cm)']]
df.columns = ['sepal_length', 'petal_length']
print(df.head())
```

3.2 2.Inspect the dataset:

```python
print(df.describe())
```

4 Step 3: Visuslize the Data 

4.1 1. Create a scatterplot to visualize the relationship:

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=df)
plt.title('Scatterplot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

5 Step 4: Correlation Analysis

5.1 1. COmputer the correlation coefficient:

```python
correlation = df.corr()
print("Correlation Matrix:\n", correlation)
```

2. Interpret the result:
> A strong positive coreelation indicates that as Swpal Length increases, Petal Length tends to increase.

6 Setep 5:Perform Linear Regression

6.1 1.Split the data into training and testing sets:

```python
x = df[['sepal_length']]
y = df['petal_length']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

6.2 2.Fit a linear regression model:

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

6.3 3.Display the regression coefficients:

```python
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
```

7 Step 6 : Assess the Fit of the Line 

7.1 1.Make prediction on the test set :

```python
y_pred = model.predict(x_test)
```

7.2 2. Compute and display metrics:

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")
```

7.3 3. Plot the regression line

```python
plt.scatter(x_test, y_test, color='blue', label='Actual values')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted values')
plt.title('Linear Regression: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```
